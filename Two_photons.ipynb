# Interferometry with two photons

### Plot the probability $P_{11}$  of a single beam splitter interferometer with two photons

```python
%matplotlib inline
from math import *
from qutip import *
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
P_11 = []
P_02 = []
P_20 = []

N = 3
a = destroy(N)

for t in arange(-pi,pi,0.1) :
    
    U = (1j*t*(tensor(a.dag(),a)+tensor(a,a.dag()))).expm() # beam splitter operator
    psi_in_vac = tensor(fock(N,0),fock(N,0))
    psi_in = tensor(a.dag(),a.dag()) * psi_in_vac
    psi_out = U*psi_in
    
    print(U)
        
    psi_11 = tensor(fock(N,1),fock(N,1))
    psi_02 = tensor(fock(N,0),fock(N,2))
    psi_20 = tensor(fock(N,2),fock(N,0))
    
    P_11.append((psi_11*psi_11.dag()*psi_out*psi_out.dag()).tr())
    P_02.append((psi_02*psi_02.dag()*psi_out*psi_out.dag()).tr())
    P_20.append((psi_20*psi_20.dag()*psi_out*psi_out.dag()).tr())

P_11_array = np.array(P_11)
P_02_array = np.array(P_02)
P_20_array = np.array(P_20)
```

```python
x = arange(-pi,pi,0.1)
plt.plot(x,P_11_array,'r',label=r'$P_{11}$')
plt.plot(x,P_02_array,'g',label=r'$P_{02}$')
plt.plot(x,P_20_array,'b',label=r'$P_{20}$')
plt.plot(x,P_11_array+P_02_array+P_20_array,'y',label='sum')
```
